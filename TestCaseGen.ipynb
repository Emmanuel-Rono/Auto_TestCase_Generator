### Processing the Extracted Data (removing mistakes), for proper utilisation in the proceeding Steps.




```python
import re

def preprocess_text(raw_text):
    # Remove excessive whitespace
    cleaned_text = re.sub(r'\s+', ' ', raw_text).strip()

    # Split requirements based on numbering or keywords like "Requirement:"
    requirements = re.split(r'(?:Requirement\s*\d*[:.-]*)', cleaned_text)

    # Remove empty strings from list
    requirements = [req.strip() for req in requirements if req.strip()]

    return requirements

# Extracted text from the document
raw_text = extract_text_from_pdf(pdf_path)

# Preprocess the text
requirements = preprocess_text(raw_text)

# Display extracted requirements
for i, req in enumerate(requirements, 1):
    print(f"Requirement {i}: {req}\n")
```

### Generate software requirements dynamically using Gemini

```python


# Function to generate software requirements dynamically using Gemini
def generate_requirements():
    prompt = """
    You are a software requirements analyst. Generate 3 software requirements for a typical web application.

    Provide requirements in the following format:
    - <Requirement 1>
    - <Requirement 2>
    - <Requirement 3>
    """

    model = genai.GenerativeModel("gemini-pro")
    response = model.generate_content(prompt)

    if response and response.text:
        return [req.strip("- ").strip() for req in response.text.strip().split("\n") if req.strip()]

    return []

# Function to generate test cases for a given requirement
def generate_test_cases(requirement):
    prompt = f"""
    You are a software testing assistant. Generate 3 test cases for the following software requirement:

    Requirement: "{requirement}"

    Provide test cases in the following format:
    1. Test Case Name: <name>
       Test Steps: <steps>
       Expected Result: <expected result>
    """

    model = genai.GenerativeModel("gemini-pro")
    response = model.generate_content(prompt)

    if response and response.text:
        test_cases = response.text.strip().split("\n\n")  # Split into individual test cases
        return [tc.strip() for tc in test_cases if tc.strip()]  # Remove empty entries

    return ["No test cases generated."]

# Step 1: Generate requirements dynamically
requirements = generate_requirements()

# Step 2: Generate test cases for the dynamically generated requirements
test_cases_dict = {}

if not requirements:
    print("No requirements generated by Gemini.")

for i, req in enumerate(requirements, 1):
    print(f"\nüîπ Generating Test Cases for Requirement {i}: {req}")
    test_cases = generate_test_cases(req)
    test_cases_dict[f"Requirement {i}"] = test_cases

    for test_case in test_cases:
        print(f"  - {test_case}")
```

## Key Step: Store extracted requirements and generated test cases in Neo4j ( This will help in informing automated Test Case Generation)


1.   Store Generated TestCases in Ne04J Auara
2.   Create the Knowledge Graph, by linking Requirements with TestCases



```python
# Function to store extracted requirements and generated test cases in Neo4j
def store_requirements_in_neo4j(requirements_with_test_cases):
    with driver.session() as session:
        for req_name, test_cases in requirements_with_test_cases.items():
            # Store the Requirement Node
            session.run("MERGE (r:Requirement {name: $name})", {"name": req_name})

            for test_case in test_cases:
                session.run("""
                MERGE (t:TestCase {name: $test_name})
                MERGE (r:Requirement {name: $req_name})
                MERGE (r)-[:VERIFIED_BY]->(t)
                """, {"test_name": test_case, "req_name": req_name})

# Store test cases in Neo4j
store_requirements_in_neo4j(test_cases_dict)
```

### Automated Test Case Generation based on Information retrieved from the Knowledge graph.


1.   Retrieve Stored Requirements from Neo4j
2.   Generate and Store New Test Cases for Existing Requirements



```python
# Function to Retrieve Stored Requirements from Neo4j
def get_stored_requirements():
    with driver.session() as session:
        result = session.run("MATCH (r:Requirement) RETURN r.name AS name")
        return [record["name"] for record in result if record["name"]]

# Function to Generate and Store New Test Cases for Existing Requirements
def generate_and_store_new_test_cases():
    stored_requirements = get_stored_requirements()
    if not stored_requirements:
        print("‚ö†Ô∏è No stored requirements found in Neo4j.")
        return

    new_test_cases_dict = {}

    for req in stored_requirements:
        test_cases = generate_test_cases(req)
        new_test_cases_dict[req] = test_cases

        with driver.session() as session:
            for test_case in test_cases:
                if not test_case.strip():
                    continue

                session.run("""
                MERGE (t:TestCase {name: $test_name})
                MERGE (r:Requirement {name: $req_name})
                MERGE (r)-[:VERIFIED_BY]->(t)
                """, {"test_name": test_case, "req_name": req})

        # Print the new test cases (not the old ones)
        print(f"\nüîπ New Test Cases for: {req}")
        for test_case in test_cases:
            print(f"  - {test_case}")

# Store Generated Test Cases in Neo4j
generate_and_store_new_test_cases()
```

```python
def validate_test_case(test_case_name):
    with driver.session() as session:
        result = session.run("MATCH (t:TestCase {name: $name}) RETURN t", name=test_case_name)
        return result.single() is not None

# Validate generated test cases
print("\nüìå Validation Results:")
for req, test_cases in test_cases_dict.items():
    print(f"\nüîπ {req}")
    for test_case in test_cases:
        is_valid = validate_test_case(test_case)
        print(f"  - {test_case}: {'‚úÖ Valid' if is_valid else '‚ùå Not Found in Knowledge Graph'}")
```

### Checking if the New Test Cases were added to the Knowledge Graph


```python
# Function to fetch and print existing test cases
def print_existing_test_cases():
    with driver.session() as session:
        # Query to retrieve test cases and their related requirements
        result = session.run("""
            MATCH (r:Requirement)-[:VERIFIED_BY]->(t:TestCase)
            RETURN r.name AS requirement_name, t.name AS test_case_name
            ORDER BY requirement_name, t.name
        """)

        # Process the query result and print it
        previous_requirement = None
        for record in result:
            requirement_name = record["requirement_name"]
            test_case_name = record["test_case_name"]

            # Print requirement header if it's new
            if requirement_name != previous_requirement:
                print(f"\nüîπ Requirement: {requirement_name}")
                previous_requirement = requirement_name

            # Print test case under the respective requirement
            print(f"  - {test_case_name}")

# Call the function to print existing test cases
print_existing_test_cases()
```

### Evaluation of the Project


1.   Retrieves AI-generated test cases from Neo4j
2.  Compares them with manually written test cases for accuracy


1.   Measures time efficiency of AI test case retrieval
2.   Checks how many requirements have at least one test case (coverage)





```python
# Function to Retrieve Stored Requirements from Neo4j
def get_stored_requirements():
  with driver.session() as session:
        result = session.run("MATCH (r:Requirement) RETURN r.name AS name")
        return [record["name"] for record in result if record["name"]]

# Function to Retrieve Stored Test Cases from Neo4j
def get_stored_test_cases():
    with driver.session() as session:
        result = session.run("""
            MATCH (r:Requirement)-[:VERIFIED_BY]->(t:TestCase)
            RETURN r.name AS requirement, collect(t.name) AS test_cases
        """)
        return {record["requirement"]: record["test_cases"] for record in result}

# Function to Evaluate Accuracy (Comparison with Manual Test Cases)
def evaluate_accuracy(ai_test_cases, manual_test_cases):
    total_ai = sum(len(cases) for cases in ai_test_cases.values())
    total_manual = sum(len(cases) for cases in manual_test_cases.values())

    if total_manual == 0:
        return 0  # Avoid division by zero

    # Simple accuracy metric: How many AI-generated test cases match the manual ones?
    common_cases = 0
    for req, ai_cases in ai_test_cases.items():
        manual_cases = set(manual_test_cases.get(req, []))
        common_cases += sum(1 for case in ai_cases if case in manual_cases)

    accuracy = (common_cases / total_manual) * 100  # Percentage match
    return round(accuracy, 2)

# Function to Measure Time Efficiency
def measure_time_efficiency():
    start_time = time.time()
    ai_test_cases = get_stored_test_cases()
    end_time = time.time()

    time_taken = round(end_time - start_time, 2)  # Execution time in seconds
    return time_taken

# Function to Measure Coverage (How many requirements have at least one test case)
def measure_coverage():
    stored_reqs = get_stored_requirements()
    stored_cases = get_stored_test_cases()

    total_requirements = len(stored_reqs)
    covered_requirements = sum(1 for req in stored_reqs if stored_cases.get(req))

    if total_requirements == 0:
        return 0  # Avoid division by zero

    coverage = (covered_requirements / total_requirements) * 100  # Percentage
    return round(coverage, 2)

#  Manual Test Cases (for Accuracy Comparison)
manual_test_cases = {
    "User Login": ["Enter valid credentials", "Enter incorrect password", "Check password reset link"],
    "Password Reset": ["Request password reset", "Use expired reset link", "Enter a weak password"]
}

# Running Evaluation Metrics
ai_test_cases = get_stored_test_cases()
accuracy = evaluate_accuracy(ai_test_cases, manual_test_cases)
time_efficiency = measure_time_efficiency()
coverage = measure_coverage()

# Display Evaluation Results
print("\n **Evaluation Report:**")
print(f" Accuracy: {accuracy}% (AI vs. Manual Test Cases)")
print(f" Time Efficiency: {time_efficiency} seconds (Data Retrieval)")
print(f" Coverage: {coverage}% (Requirements with Test Cases)")

# Close Neo4j Connection
driver.close()
```
